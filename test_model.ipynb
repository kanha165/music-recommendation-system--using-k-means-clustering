```python
import pickle
import pandas as pd

with open("song_recommender.pkl", "rb") as f:
    model_data = pickle.load(f)

df = model_data["df"]
similarity = model_data["similarity"]
```

```python
def recommend(song_name, top_n=5):
    if song_name not in df['song_name'].values:
        return [" Song not found in database"]

    index = df[df['song_name'] == song_name].index[0]
    similarity_scores = list(enumerate(similarity[index]))

    similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)
    similarity_scores = similarity_scores[1:top_n+1]

    recommended_songs = []

    for i in similarity_scores:
        song = df.iloc[i[0]]
        recommended_songs.append(
            f"{song['song_name']} — {song['artist_name']}"
        )

    return recommended_songs
```

```python
while True:
    song_input = input("Enter a song name (or 'exit'): ")

    if song_input.lower() == "exit":
        break

    results = recommend(song_input)

    print("\n Recommended Songs:")
    for r in results:
        print(">>", r)
```
